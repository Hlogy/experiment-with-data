```python
import pandas as pd
```

```python
data = pd.read_excel('pwt.xlsx', sheet_name ='Data')
```

```python
# deleting the countrycode column
data.drop(columns=['countrycode'], inplace = True)
data
```

```python
# delete years from 1950 to 1999
indecies = (data[ data['year']<= 1999].index)
data.drop(indecies, inplace = True)
```

```python
data
```

```python
# deleting rows with 4 or more NaN
num_of_nan = data.isnull().sum(axis=1)
index_of_nan = num_of_nan.index

for i in index_of_nan:
#     print(i)
    if num_of_nan[i]>=4:
        data.drop([i], inplace = True)
        
data
```

```python
# setting the country and year columns as multiIndex
data.set_index(['country', 'year'])
```

```python
data.to_excel("pwt_clean.xlsx") 
```
