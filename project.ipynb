```python
import pandas as pd
import numpy as np # linear algebra
import matplotlib.pyplot as plt
import math
```

```python
#load the cleaned datasets of atm, pwt, and gini
atm = pd.read_excel('atm_clean.xlsx')
pwt = pd.read_excel('pwt_clean.xlsx')
gini = pd.read_excel('gini_clean.xlsx')
```

```python
gini.drop(gini.columns[[0]], axis=1, inplace = True)
```

```python
gini_country = gini.query('country=="China" | country=="Pakistan" | country=="Turkey" | country=="Nigeria" | country=="Hungary" \
                | country=="S.A" | country=="Poland" | country=="Romania" | country=="Mexico" | country=="Brazil"')
```

```python
gini_T = gini_country.transpose()
```

```python
gini_T.reset_index(inplace = True)
```

```python
gini_T.columns = ['year','Brazil','China','Hungary','Mexico','Nigeria','Pakistan','Poland','Romania','S.A','Turkey']
```

```python
gini_T = gini_T.drop([gini_T.index[0]])
gini_T.rename(columns = {'index':'year'}, inplace = True) 
```

```python
gini_T
```

```python
# plot gini data for the 10 countries for years 2004-2015
ax = plt.gca()
plt.xticks([2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013 , 2014 , 2015])
plt.yticks(np.arange(0.2, 0.65, 0.025).tolist())

gini_T.plot(kind='line',x='year',y='Brazil',ax=ax)
gini_T.plot(kind='line',x='year',y='China', color='red', ax=ax)
gini_T.plot(kind='line',x='year',y='Hungary', color='#99ff33', ax=ax)
gini_T.plot(kind='line',x='year',y='Mexico', color='#cc33ff', ax=ax)
gini_T.plot(kind='line',x='year',y='Nigeria', color='#666699', ax=ax)
gini_T.plot(kind='line',x='year',y='Pakistan', color='#339966', ax=ax)
gini_T.plot(kind='line',x='year',y='Poland', color='#ff9900', ax=ax)
gini_T.plot(kind='line',x='year',y='Romania', color='#993300', ax=ax)
gini_T.plot(kind='line',x='year',y='S.A', color='#800000', ax=ax)
gini_T.plot(kind='line',x='year',y='Turkey', color='#5e8000', ax=ax)

plt.xlabel("Years")
plt.ylabel("Gini Value")

fig = plt.gcf()
fig.set_size_inches(15, 8, forward=True)
fig.suptitle('Gini Values from 2004 to 2015', fontsize=20)


plt.show()
```
