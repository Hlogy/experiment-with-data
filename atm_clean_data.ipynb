```python
import pandas as pd
```

```python
data = pd.read_excel('atm.xls')
```

```python
data
# data.loc[:,'Unnamed: 2']
data.columns=["C "+str(i) for i in range(1,65)]
```

```python
data
```

```python
data.drop(data.iloc[:, 4:48], axis = 1, inplace = True) 
data
```

```python
#rename columns with number of years from 2000 to 2019, 2020 is not included but is needed for the range/ next step we'll remove it for the first two columns
data.columns=[""+str(i) for i in range(2000, 2020)]
# rename the 2000 and 2001 columns accordingly:
data = data.rename(columns={'2000': 'Country Name', '2001': 'Country Code'}, index={'ONE': 'one'})
# delete 2002 and 2003 and 2019 columns
data.drop(columns=['2002', '2003', '2019'], inplace = True)
# delete rows 0,1,2
data.drop([0,1,2], inplace = True)
```

```python
data
```

```python
# count the number of NaN and save it in variable num_of_nan
num_of_nan = data.isnull().sum(axis=1)
num_of_nan
```

```python
# use for loop to go through rows where number of NaN is 4 or more, and delete that row from Data
for i in range(3,267):
    if num_of_nan.loc[i] >= 4:
        data.drop([i], inplace = True)

data
```

```python
# set country name as data index
data.set_index('Country Name')
```

```python
data.to_excel("atm_clean.xlsx") 
```
